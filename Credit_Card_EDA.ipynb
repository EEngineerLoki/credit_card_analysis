```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Data Reading Process

```python
data = pd.read_csv(r'data/credit_card_transactions.csv')
```

```python
data.shape
```

### Data Cleaning and Filtering

```python
cols = ['Unnamed: 0', 'lat', 'long', 'unix_time', 'merch_lat', 'merch_long', 'is_fraud', 'merch_zipcode', 'category']
data.drop(cols, axis=1, inplace=True)
```

```python
data.rename(columns={
    'trans_date_trans_time' : 'transaction_date',
    'cc_num' : 'card_number',
    'amt' : 'purchased_amount',
    'city_pop' : 'city_population',
    'dob' : 'date_of_birth',
    'trans_num' : 'transaction_number'
}, inplace=True)
```

```python
data['transaction_date'] = pd.to_datetime(data['transaction_date'], dayfirst=True)
data['date_of_birth'] = pd.to_datetime(data['date_of_birth'], dayfirst=True)
```

```python
txt = ['merchant', 'first', 'last', 'gender', 'street', 'city', 'job', 'transaction_number', 'state', 'card_number']
```

```python
data[txt] = data[txt].astype('string')
```

```python
data.info()
```

```python
data['merchant'] = data['merchant'].str.replace('fraud_', '')
```

```python
data['full_name'] = data['first'] + ' ' + data['last']
name = data.pop('full_name')
data.insert(5, 'full_name', name)
data.drop(['first', 'last'], inplace=True, axis=1)
```

```python
data.head()
```

##### Gender Based Spending

```python
gTransactions = data.groupby(['transaction_date', 'gender'], as_index=False).agg({
    'transaction_number' : 'count',
    'purchased_amount' : 'sum'
}).rename(columns={'transaction_number' : 'number_of_transactions', 'purchased_amount' : 'total_amount'})
gTransactions['Men'] = gTransactions[gTransactions.gender == 'M']['number_of_transactions']
gTransactions['Women'] = gTransactions[gTransactions.gender == 'F']['number_of_transactions']
gTsorted = gTransactions.groupby('transaction_date', as_index=False)[['Men', 'Women']].sum()
gTsorted.set_index('transaction_date', inplace=True)
gTsorted
```

```python
gTransactions
```

```python
sns.set_theme()

# Plot sepal width as a function of sepal_length across days
g = sns.lmplot(
    data=gTransactions,
    x="number_of_transactions", y="total_amount", hue="gender",
    height=5
).set(title='Bivariate Analysis by Gender')

# Use more informative axis labels than are provided by default
g.set_axis_labels("Transaction Volume (n)", "Amount Spent (Dollars)")
```

```python
sns.set_theme(style="ticks")

g = sns.jointplot(data= gTransactions, x='number_of_transactions', y='total_amount', kind="hex", color="#4CB391")
g.set_axis_labels("Transaction Volume (n)", "Amount Spent (Dollars)")
```

```python
gTsorted.mean()
```

```python
gTsorted.loc['2019-12-02' : '2020-01-01'][gTsorted > 2500].fillna(0)
```

```python
gTsorted.loc['2019-12-01' : '2020-01-01'][gTsorted > 2500].fillna(0).plot(figsize=(18,8)).set(title='Tracking of Above Mean Spending in the Most Active Month')
```

```python
gAmount = data.groupby(['transaction_date', 'gender'], as_index=False)['purchased_amount'].agg([
    ('total_amount', 'count')
])
gAmount['Men'] = gAmount[gAmount.gender == 'M'][['total_amount']]
gAmount['Women'] = gAmount[gAmount.gender == 'F'][['total_amount']]
gAsorted = gAmount.groupby('transaction_date', as_index=False)[['Men', 'Women']].sum()
gAsorted.set_index('transaction_date', inplace=True)
gAsorted
```

```python
gAsorted.loc[: '2020-01-01'].plot(figsize=(18,8)).set(title='Correlation of Spending of Genders in a Year')
```

##### Location Mapping of Individual Spendings

```python
location = data.loc[:, ['transaction_date', 'full_name', 'street', 'city', 'state', 'city_population', 'card_number', 'transaction_number', 'purchased_amount']]
location.head()
```

```python
location['full_address'] = location[['street', 'city', 'state']].agg(', '.join, axis=1)
```

```python
address = location.pop('full_address')
location.insert(2, 'full_address', address)
location
```

```python
location.groupby(['full_name', 'full_address'], as_index=False)['card_number'].agg([
    ('cards_owned', 'nunique')
]).sort_values(by='cards_owned', ascending=False)
```

```python
location.groupby(['full_address'], as_index=False)['transaction_number'].nunique().sort_values(by='transaction_number', ascending=False)
```

```python
topLpurchase = location.groupby(['full_address'], as_index=False)['transaction_number'].nunique().sort_values(by='transaction_number', ascending=False).iloc[:10]
topLpurchase
```

```python
topLamount = location.groupby(['full_address'], as_index=False)['purchased_amount'].sum().sort_values(by='purchased_amount', ascending=False).iloc[:10]
topLamount
```

```python
topLocation = topLpurchase[topLpurchase.full_address.isin(topLamount.full_address)]
topLocation
```

```python
topLcustomers = location[location.full_address.isin(topLocation.full_address)][['full_name', 'full_address', 'transaction_number', 'purchased_amount']]
topLocCus = topLcustomers.groupby(['full_name', 'full_address'], as_index=False).agg({
    'transaction_number' : 'count',
    'purchased_amount' : 'sum'
}).rename(columns={'transaction_number' : 'number_of_transactions', 'purchased_amount' : 'total_amount'})
topLocCus
```

##### Credit Card Owners to City Population Ratio

```python
cardOwners = location.groupby(['city', 'city_population'], as_index=False)['full_name'].agg([
    ('card_owners_in_city', 'nunique')
]).sort_values(by='card_owners_in_city', ascending=False)
cardOwners['owners:city_pop ratio'] = (cardOwners.card_owners_in_city / cardOwners.city_population) * 100
cardOwners
```

```python
owners = location.groupby(['transaction_date', 'city'], as_index=False).agg({
    'purchased_amount' : 'sum',
    'transaction_number' : 'count'
}).rename(columns={'purchased_amount' : 'total_amount', 'transaction_number' : 'number_of_transactions'}).sort_values(by='total_amount', ascending=False)
```

```python
cardCityAmount = pd.pivot_table(owners, index='transaction_date', columns='city', values='total_amount').fillna(0)
cardCityAmount.plot(figsize=(12,6), legend=None).set(title='Graphical Representation of Amount Spending by City')
```

```python
cardCityAmount = pd.pivot_table(owners, index='transaction_date', columns='city', values='number_of_transactions').fillna(0)
cardCityAmount.plot(figsize=(12,6), legend=None).set(title='Graphical Representation of Transaction Volume by City')
```

```python
import matplotlib as mpl
sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)

sns.histplot(
    owners,
    x="total_amount",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5,
    log_scale=True,
    kde=True
).set(title='Normally Distributed Graph by Amount Spending and Occurences')
ax.xaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
ax.set_xticks([500, 1000, 2000, 5000, 10000])
```

##### Listing of Usage of Credit Card According to Prefered Merchants

```python
merchants = pd.merge(data, location, how='left')
merch = merchants.pop('full_address')
merchants.insert(6, 'full_address', merch)
merchants = merchants.loc[:,['transaction_date', 'full_name', 'full_address', 'merchant', 'transaction_number', 'purchased_amount']]
merchants.head()
```

```python
days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
merchants['day'] = merchants['transaction_date'].dt.day_name()
merchants['day'] = pd.Categorical(merchants['day'], categories=days, ordered=True)
merchDay = merchants.groupby(['day', 'transaction_date']).agg({
    'transaction_number' : 'count',
    'purchased_amount' : 'sum'
})
merchDay
```

```python
sns.set_theme(style="whitegrid")

cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=merchDay,
    x="transaction_number", y="purchased_amount",
    hue="day", 
    palette="ch:s=.25,rot=-.25", sizes=(10, 200),
).set(title='Bivariate Analysis by the Days of the Week')
g.set_axis_labels("Transaction Volume (n)", "Amount Spent (Dollars)")
g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

##### Sorting Spending Indications by Age Categories

```python
data = pd.merge(data, location, how='left')
address = data.pop('full_address')
data.insert(6, 'full_address', address)
data.head()
```

```python
ageData = data.loc[:,['transaction_date', 'full_name', 'full_address', 'date_of_birth', 'merchant', 'transaction_number', 'purchased_amount']]
ageData
```

```python
ageData['age'] = (ageData['transaction_date'] - ageData['date_of_birth']) / 365
ageData['age'] = ageData['age'].dt.days.astype('int16')
ageData
```

```python
ageData['age_category'] = pd.cut(x=ageData['age'], bins=[0,10,20,30,40,100], labels=['infant', 'teens', 'young_aduls', 'middle_aged', 'old']) 
ageData
```

```python
ageGroup = ageData.groupby('age_category', as_index=False).agg({
    'transaction_number' : 'count',
    'purchased_amount' : 'sum'
}).rename(columns={'transaction_number' : 'number_of_transactions', 'purchased_amount' : 'total_amount'}).sort_values(by='total_amount', ascending=False)
ageGroup
```

```python
ageData.merchant.nunique()
```

```python
ageData.groupby(['age_category', 'merchant'], as_index=False)['purchased_amount'].sum().sort_values(by='purchased_amount', ascending=False).iloc[:10]
```

```python
ageTotal = ageData.groupby(['age'], as_index=False).agg({
    'transaction_number' : 'count',
    'purchased_amount' : 'sum'
}).rename(columns={'transaction_number' : 'number_of_transactions', 'purchased_amount' : 'total_amount'}).sort_values(by='total_amount', ascending=False)
ageTotal['age_category'] = pd.cut(x=ageTotal['age'], bins=[0,10,20,30,40,100], labels=['infant', 'teens', 'young_adult', 'middle_aged', 'old'])
ageTotal.sort_values(by='age')
```

```python
sns.set_theme(style="darkgrid")
g = sns.relplot(data=ageTotal, 
            x='number_of_transactions', 
            y='total_amount', 
            hue='age_category',
            palette="ch:s=.25,rot=-.25").set(title='Bivariate Analysis by Age Category')
g.set_axis_labels("Transaction Volume (n)", "Amount Spent (Dollars)")
```

```python
g = sns.displot(
    data=ageTotal,
    x="number_of_transactions", hue="age_category",
    kind="kde", height=6,
    multiple="fill", clip=(0, None),
    palette="ch:rot=-.25,hue=1,light=.75",
).set(title='Density of Transaction Volume by Age Category')
g.set_axis_labels("Transaction Volume (n)")
```

```python
g = sns.displot(
    data=ageTotal,
    x="total_amount", hue="age_category",
    kind="kde", height=6,
    multiple="fill", clip=(0, None),
    palette="ch:rot=-.25,hue=1,light=.75",
).set(title='Density of Spending by Age Category')
g.set_axis_labels("Amount Spent (Dollars)")
```

##### Days and Months with Most Transactions Taken

```python
data.head()
```

```python
day = data.loc[:,['transaction_date', 'purchased_amount', 'transaction_number']]
day['day'] = day['transaction_date'].dt.day_name()
day['month'] = day['transaction_date'].dt.month_name()
days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
day['day'] = pd.Categorical(day['day'], categories=days, ordered=True)
day['month'] = pd.Categorical(day['month'], categories=months, ordered=True)
dayTransaction = day.groupby(['day', 'month'], as_index=False).agg({
    'transaction_number' : 'count',
    'purchased_amount' : 'sum'
})
dayTransaction
```

```python
g = sns.displot(
    data=dayTransaction,
    x="transaction_number", hue="month",
    kind="kde", height=6,
    multiple="fill", clip=(0, None),
    palette="ch:rot=-.25,hue=1,light=.75",
).set(title='Density of Monthly Transactions')
g.set_axis_labels("Transaction Volume (n)")
```

```python
g = sns.displot(
    data=dayTransaction,
    x="transaction_number", hue="day",
    kind="kde", height=6,
    multiple="fill", clip=(0, None),
    palette="ch:rot=-.25,hue=1,light=.75",
).set(title='Density of Transactions based on the Days of the Week')
g.set_axis_labels("Transaction Volume (n)")
```

```python
dayTpivot = pd.pivot_table(dayTransaction, index='day', columns='month', values='transaction_number')
dayTpivot
```

```python
sns.set_theme()

f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(dayTpivot, annot=True, linewidths=.5, ax=ax).set(title='Volume of Transactions based on the Days and Months')
```
